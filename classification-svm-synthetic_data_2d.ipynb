```python
import numpy as np
import matplotlib.pyplot as plt

from itertools import product

from sklearn.svm import SVC
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
def train_and_analyze_svm(X, Y, kernel='rbf', C=1.0, gamma='scale', train_size=0.2, random_state=42):

    # Split into training and testing sets
    X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=train_size, random_state=random_state)

    # Train an SVM model
    svm = SVC(kernel=kernel, C=C, gamma=gamma, random_state=random_state)
    svm.fit(X_train, Y_train)

    # Make predictions for the test set
    Y_pred = svm.predict(X_test)

    # Evaluate the SVM performance
    print("Accuracy:", accuracy_score(Y_test, Y_pred))
    print("Confusion Matrix:\n", confusion_matrix(Y_test, Y_pred))
    print("Classification Report:\n", classification_report(Y_test, Y_pred))

    # Build a 2D grid
    x_min, x_max = 1.2 * X[:, 0].min(), 1.2 * X[:, 0].max() 
    y_min, y_max = 1.2 * X[:, 1].min(), 1.2 * X[:, 1].max() 
    grid_x, grid_y = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
    X_grid = np.c_[grid_x.ravel(), grid_y.ravel()] # stack the two flattened arrays column-wise

    # Make SVM prediction for each grid point
    Y_svm = svm.predict(X_grid) 

    # Draw the decision boundary and the input dataset    
    Y_svm = Y_svm.reshape(grid_x.shape)
    plt.figure(figsize=(8,5))
    plt.contourf(grid_x, grid_y, Y_svm, alpha=0.3, cmap=plt.cm.coolwarm)
    plt.scatter(X_train[:, 0], X_train[:, 1], c=Y_train, cmap=plt.cm.coolwarm, edgecolors='k')
    plt.scatter(X_test [:, 0], X_test [:, 1], c=Y_test,  cmap=plt.cm.coolwarm, edgecolors='k')
    plt.xlabel("x1")
    plt.ylabel("x2")
    plt.title("SVM decision boundary")
    plt.show()
```

```python
# Inputs

N = 500 # number of samples per syntetic dataset
kernel = 'rbf' #
C = 1.0 #
gamma = 'scale' #
ts = 0.2 # train size

class_sep_list = [0.5, 1.0, 2.0] #
n_clusters_per_class_list = [1, 2] #
random_state_list = [35, 40, 45] #

# Loop over all combinations of class separation values, random states, and number of clusters per class
for cs, nc, rs in product(class_sep_list, n_clusters_per_class_list, random_state_list):

    # Build a synthetic dataset
    print(f"Building synthetic 2D dataset with {N} samples using: " 
          f"class_sep = {cs}, n_clusters_per_class = {nc}, random_state = {rs}")
    X, Y = make_classification(
        n_samples=N, n_features=2, class_sep=cs, 
        n_informative=2, n_redundant=0, n_clusters_per_class=nc, random_state=rs)

    # Run an SVM and analyze its performance
    print(f"Running SVM with: " 
          f"kernel = {kernel}, C = {C}, gamma={gamma}")    
    train_and_analyze_svm(
        X, Y, kernel=kernel, C=C, gamma=gamma, train_size=ts, random_state=rs)
```
