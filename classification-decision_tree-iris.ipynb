```python
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import tree
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from IPython.display import display, HTML
```

```python
# Get the IRIS dataset
iris = load_iris()
X, Y = iris.data, iris.target
print(f"Shapes of IRIS dataset arrays - Features array: {X.shape}, Labels array: {Y.shape}")
```

```python
# Convert to DataFrame and print dataset
df = pd.DataFrame(X, columns=iris.feature_names)
df['species'] = Y  # Add target labels
# Convert DataFrame to an interactive HTML table with scrollable output
html = df.to_html(max_rows=20, max_cols=10)
display(HTML('<style> .dataframe {max-height: 300px; overflow-y: scroll; display: block; } </style>'))
display(HTML(df.to_html()))
# display(df) # simpler option if the above doesn't work
```

```python
# Split IRIS dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# Create and train a decision tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, Y_train)
```

```python
# Draw the tree
plt.figure(figsize=(12,8))  # Increase figure size
tree.plot_tree(clf, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```

```python
# Draw the tree
Y_pred = clf.predict(X_test)
```

```python
# Calculate the accuracy
accuracy = accuracy_score(Y_test, Y_pred)
print(f'Accuracy: {100*accuracy:.2f}%')
```
